```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#  Student and Faculty Demographic CSV FILES
path = "student_demo.csv"

# Read the data
demo = pd.read_csv(path)
```

```python
demo['completion rate in 4 years (asian)'] = demo['completion rate in 4 years (asian)'] * 100
demo['completion rate in 4 years (white)'] = demo['completion rate in 4 years (white)'] * 100
demo['completion rate in 4 years (black)'] = demo['completion rate in 4 years (black)'] * 100
demo['completion rate in 4 years (american indian)'] = demo['completion rate in 4 years (american indian)'] * 100
demo['completion rate in 4 years (hawaiian or pacific islander)'] = demo['completion rate in 4 years (hawaiian or pacific islander)'] * 100
demo['completion rate in 4 years (hispanic)'] = demo['completion rate in 4 years (hispanic)'] * 100
demo['faculty demographic (asian)'] = demo['faculty demographic (asian)'] * 100
demo['faculty demographic (white)'] = demo['faculty demographic (white)'] * 100
demo['faculty demographic (black)'] = demo['faculty demographic (black)'] * 100
demo['faculty demographic (american indian)'] = demo['faculty demographic (american indian)'] * 100
demo['faculty demographic (hawaiian or pacific islander)'] = demo['faculty demographic (hawaiian or pacific islander)'] * 100
demo['faculty demographic (hispanic)'] = demo['faculty demographic (hispanic)'] * 100
demo['pell grant rate'] = demo['pell grant rate'] * 100
demo
```

```python
demo = demo[demo['completion rate in 4 years (asian)'] != 0]
```

```python
demo.columns
```

```python
# ASIAN STUDENTS AND ASIAN FACULTY
# Set the figure size before creating the bar chart
plt.figure(figsize=(12, 6))
plt.title('Faculty Diversity and Graduation Rates for Minority Students')
plt.xlabel('Completion Rate in 4 years for Asian Students')
plt.ylabel('Faculty Race/Ethnicity (Asian)')

x_axis = demo["completion rate in 4 years (asian)"]
y_axis = demo["faculty demographic (asian)"]

scatter_plot = plt.scatter(x_axis, y_axis, c=demo['pell grant rate'])
plt.xticks(rotation=90)

plt.colorbar(scatter_plot, label='Pell Grant Rate')

plt.tight_layout()
plt.show()
```

```python
demo = demo[demo['completion rate in 4 years (white)'] != 0]
```

```python
# WHITE STUDENTS AND WHITE FACULTY
# Set the figure size before creating the bar chart
plt.figure(figsize=(12, 6))
plt.title('Faculty Diversity and Graduation Rates for Minority Students')
plt.xlabel('Completion Rate in 4 years for White Students')
plt.ylabel('Faculty Race/Ethnicity (White)')

x_axis = demo["completion rate in 4 years (white)"]
y_axis = demo["faculty demographic (white)"]

plt.scatter(x_axis, y_axis, c=demo['pell grant rate'])

# Plotting the regression line
linear_black = st.linregress(x_axis, y_axis)
x_min = min(x_axis)
x_max = max(x_axis)
y_min = linear_black.slope * x_min + linear_black.intercept
y_max = linear_black.slope * x_max + linear_black.intercept
plt.plot([x_min, x_max], [y_min, y_max], color='red', label='Linear Regression')
plt.colorbar(scatter_plot, label='Pell Grant Rate')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
